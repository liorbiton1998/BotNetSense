```python
# Load the full dataset
nov22 = pd.read_csv("/content/drive/MyDrive/Thesis/classification_datasets_cleaned/tweets_election_nov2022_bot_vs_random.csv")

# Create shortened version of text (optional, for LLM context size)
nov22["short_text"] = nov22["clean_text"].str.slice(0, 300)

# Apply political keyword filter
filtered_df = nov22[nov22["short_text"].apply(likely_political)].reset_index(drop=True)

print(f"Filtered down to {len(filtered_df)} tweets likely related to politics")
```

```python
def create_prompts_hebrew_for_bot_detection(df, avg_df, election_name):
    network_cols = [
        'in_degree', 'out_degree', 'community_size',
        'bots_neighbors', 'non_bots_neighbors',
        'betweenness_centrality', 'closeness_centrality'
    ]

    rtl = '\u202B'  # Unicode RTL override

    # df = df.dropna(subset=network_cols + ['clean_text']).copy()
    df = df.copy()

    # Get average values for this election
    avg_row = avg_df[avg_df['election'] == election_name].iloc[0]

    # # Normalize each column by the average
    # for col in network_cols:
    #     df[col + '_rel'] = df[col] / (avg_row[col] + 1e-8)  # avoid division by zero
    # Only normalize centrality measures
    df['betweenness_centrality_rel'] = df['betweenness_centrality'] / (avg_row['betweenness_centrality'] + 1e-8)
    df['closeness_centrality_rel'] = df['closeness_centrality'] / (avg_row['closeness_centrality'] + 1e-8)


    prompts = []

    for _, row in df.iterrows():
        tweet_id = row['tweet_id']
        user_id = row['user_id']
        tweet_text = row['clean_text']
        is_bot = row['is_bot']

        in_d = row['in_degree']
        out_d = row['out_degree']
        btw = row['betweenness_centrality_rel']
        clo = row['closeness_centrality_rel']
        comm_size = row['community_size']
        bot_neighbors = row['bots_neighbors']
        non_bot_neighbors = row['non_bots_neighbors']

        # Prompts
        #context_line = "×›×œ ×”××“×“×™× ××ª××¨×™× ××ª ×”×§×•×“×§×•×“ ×‘×™×—×¡ ×œ×§×•×“×§×•×“ ×××•×¦×¢ ×‘×¨×©×ª"
        context_line = (
            "××ª×” ××•××—×” ×‘×–×™×”×•×™ ×‘×•×˜×™× ×•×”×ª×¢×¨×‘×•×ª ×–×¨×”. "
            "×—×œ×§ ××”×¦×™×•×¦×™× ×”× ×©×œ ×‘×•×˜×™× â€” ×œ×¨×•×‘ ××§×˜×‘×™× ×•×§×™×¦×•× ×™×™×, ××©×ª×™×™×›×™× ×œ×¨×©×ª×•×ª ×§×˜× ×•×ª ××• ××‘×•×“×“×•×ª ×©×œ× ××¦×œ×™×—×™× ×œ×”×™×˜××¢ ×‘×¨×©×ª. ××—×¨×™× ×©×œ ××©×ª××©×™× ×××™×ª×™×™×."
        )

        minimal = rtl + f"""{context_line}
        ×¦×™×•×¥: "{tweet_text}"
        ×”×× ×”×¦×™×•×¥ × ×›×ª×‘ ×¢×œ ×™×“×™ ×‘×•×˜? ×¢× ×” ×¨×§ ×‘×›×Ÿ/×œ×
        """

        minimal_cot = rtl + f"""{context_line}
        ×¦×™×•×¥: "{tweet_text}"
        ×”×× ×”×¦×™×•×¥ × ×›×ª×‘ ×¢×œ ×™×“×™ ×‘×•×˜? ×¢× ×” ×›×Ÿ/×œ× ×•×”×¡×‘×¨
        """

        degree = rtl + f"""{context_line}
        ×¦×™×•×¥: "{tweet_text}"
        - ×¨×™×˜×•×•×˜×™× ×©×§×™×‘×œ ×”××©×ª××©: {in_d:.2f}
        - ×¨×™×˜×•×•×˜×™× ×©×‘×™×¦×¢ ×”××©×ª××©: {out_d:.2f}
        ×‘×”×ª×‘×¡×¡ ×¢×œ ×”×¦×™×•×¥ ×•× ×ª×•× ×™ ×”×¨×©×ª, ×”×× ×”×¦×™×•×¥ × ×›×ª×‘ ×¢×œ ×™×“×™ ×‘×•×˜? ×¢× ×” ×¨×§ ×‘×›×Ÿ/×œ×
        """

        centrality = rtl + f"""{context_line}
        ×¦×™×•×¥: "{tweet_text}"
        - ××¨×›×–×™×•×ª ×‘×™×Ÿ-×¦××ª×™×ª ×‘×™×—×¡ ×œ×××•×¦×¢: {btw:.2f}
        - ××¨×›×–×™×•×ª ×§×¨×‘×” ×‘×™×—×¡ ×œ×××•×¦×¢: {clo:.2f}
        ×‘×”×ª×‘×¡×¡ ×¢×œ ×”×¦×™×•×¥ ×•×”××¨×›×–×™×•×ª ×‘×¨×©×ª, ×”×× ×”×¦×™×•×¥ × ×›×ª×‘ ×¢×œ ×™×“×™ ×‘×•×˜? ×¢× ×” ×¨×§ ×‘×›×Ÿ/×œ×
        """

        community = rtl + f"""{context_line}
        ×¦×™×•×¥: "{tweet_text}"
        - ×’×•×“×œ ×”×§×”×™×œ×”: {comm_size:.2f}
        - ×©×›× ×™× ×‘×•×˜×™×: {bot_neighbors:.2f}
        - ×©×›× ×™× ×œ×-×‘×•×˜×™×: {non_bot_neighbors:.2f}
        ×‘×”×ª×‘×¡×¡ ×¢×œ ×”×¦×™×•×¥ ×•×”×§×©×¨×™× ×”×—×‘×¨×ª×™×™×, ×”×× ×”×¦×™×•×¥ × ×›×ª×‘ ×¢×œ ×™×“×™ ×‘×•×˜? ×¢× ×” ×¨×§ ×‘×›×Ÿ/×œ×
        """

        full = rtl + f"""{context_line}
        ×¦×™×•×¥: "{tweet_text}"
        - ×¨×™×˜×•×•×˜×™× ×©×§×™×‘×œ: {in_d:.2f}
        - ×¨×™×˜×•×•×˜×™× ×©×‘×™×¦×¢: {out_d:.2f}
        - ××¨×›×–×™×•×ª ×‘×™×Ÿ-×¦××ª×™×ª ×‘×™×—×¡ ×œ×××•×¦×¢: {btw:.2f}
        - ××¨×›×–×™×•×ª ×§×¨×‘×” ×‘×™×—×¡ ×œ×××•×¦×¢: {clo:.2f}
        - ×’×•×“×œ ×”×§×”×™×œ×”: {comm_size:.2f}
        - ×©×›× ×™× ×‘×•×˜×™×: {bot_neighbors:.2f}
        - ×©×›× ×™× ×œ×-×‘×•×˜×™×: {non_bot_neighbors:.2f}
        ×‘×”×ª×‘×¡×¡ ×¢×œ ×”×¦×™×•×¥ ×•×›×œ×œ × ×ª×•× ×™ ×”×¨×©×ª, ×”×× ×”×¦×™×•×¥ × ×›×ª×‘ ×¢×œ ×™×“×™ ×‘×•×˜? ×¢× ×” ×¨×§ ×‘×›×Ÿ/×œ×
        """

        cot = rtl + f"""{context_line}
        ×¦×™×•×¥: "{tweet_text}"
        - ×¨×™×˜×•×•×˜×™× ×©×§×™×‘×œ: {in_d:.2f}
        - ×¨×™×˜×•×•×˜×™× ×©×‘×™×¦×¢: {out_d:.2f}
        - ××¨×›×–×™×•×ª ×‘×™×Ÿ-×¦××ª×™×ª ×‘×™×—×¡ ×œ×××•×¦×¢: {btw:.2f}
        - ××¨×›×–×™×•×ª ×§×¨×‘×” ×‘×™×—×¡ ×œ×××•×¦×¢: {clo:.2f}
        - ×’×•×“×œ ×”×§×”×™×œ×”: {comm_size:.2f}
        - ×©×›× ×™× ×‘×•×˜×™×: {bot_neighbors:.2f}
        - ×©×›× ×™× ×œ×-×‘×•×˜×™×: {non_bot_neighbors:.2f}
        ×‘×”×ª×‘×¡×¡ ×¢×œ ×”×¦×™×•×¥ ×•×›×œ×œ × ×ª×•× ×™ ×”×¨×©×ª, ×”×× ×”×¦×™×•×¥ × ×›×ª×‘ ×¢×œ ×™×“×™ ×‘×•×˜? ×¢× ×” ×›×Ÿ/×œ× ×•×”×¡×‘×¨
        """

        prompts.append({
            'tweet_id': tweet_id,
            'user_id': user_id,
            'is_bot': is_bot,
            'Minimal Prompt': minimal,
            'Minimal Prompt with CoT': minimal_cot,
            'Degree-Based Prompt': degree,
            'Centrality-Based Prompt': centrality,
            'Community Prompt': community,
            'Full-Feature Prompt': full,
            'Chain-of-Thought Prompt': cot
        })

    return pd.DataFrame(prompts)
```

```python
# Load data
avg_df = pd.read_csv("/content/drive/MyDrive/Thesis/all_averages_summary.csv")

# Set election name to match what's in avg_df
election_name = "tweets_election_nov2022_processed"

# Create prompts
prompts_df = create_prompts_hebrew_for_bot_detection(balanced_df, avg_df, election_name)

# Separate bots and non-bots
bots = prompts_df[prompts_df["is_bot"] == 1].reset_index(drop=True)
non_bots = prompts_df[prompts_df["is_bot"] == 0].reset_index(drop=True)

# Limit to the shorter length to ensure equal pairing
min_len = min(len(bots), len(non_bots))
bots = bots.iloc[:min_len]
non_bots = non_bots.iloc[:min_len]

# Interleave rows: bot, non-bot, bot, non-bot...
interleaved_rows = [row for pair in zip(bots.iterrows(), non_bots.iterrows()) for row in pair]
interleaved_df = pd.DataFrame([row[1] for row in interleaved_rows])
```
